```python
import pandas as pd
crop_data = pd.read_csv('crop_production_main.csv')
rain_data = pd.read_csv('rainfall_data.csv')
check = pd.read_csv('check.csv')
crop_data.head()
```

```python
crop_data.drop(crop_data.loc[crop_data['year']==1997].index, inplace=True)
crop_data.drop(crop_data.loc[crop_data['year']==1998].index, inplace=True)
crop_data.drop(crop_data.loc[crop_data['year']==1999].index, inplace=True)
crop_data.to_csv('crop_production_main.csv')
```

```python
crop_data.drop(crop_data.loc[crop_data['state'] == 'puducherry'].index, inplace=True)
crop_data.to_csv('crop_production_main.csv')
```

```python
df = pd.merge(crop_data, rain_data, how="left", right_on=["SUBDIVISION", "YEAR"], left_on=["state", "year"])
df.to_csv('check.csv', index=False);
```
